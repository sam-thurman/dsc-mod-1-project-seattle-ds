# Findings

The goal of this project was to follow up on the 2016 report of Opportunity Youth in Southern King County. Opportunity Youth are defined as people aging between 16-24 who are neither working, nor enrolled in school.  We analyzed samples of census data in relevant areas of King County to better understand who makes up this population, and hypothesize how city programs can better support these groups.  The purpose of this notebook is to serve as a visual representation of our findings.
